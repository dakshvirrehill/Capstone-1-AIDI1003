```python
sig = sigmoid_kernel(dfnor, dfnor)
def recommend(name, sig=sig):
    index = names[name]
    sScore = list(enumerate(sig[index]))
    sScore = sorted(sScore, key=lambda x: x[1], reverse=True)
    sScore = sScore[0:11]
    songIndex = [i[0] for i in sScore]
    return df['name'].iloc[songIndex]
```
